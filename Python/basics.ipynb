## 1. Python Syntax

- Python uses **indentation** (whitespace) to define code blocks, not curly braces `{}`.
- Statements usually end at a newline (no semicolon needed).
- Comments start with `#`.
- Indentation must be consistent (4 spaces is the convention).

````python
# This is a comment
if 5 > 2:
    print("Five is greater than two!")  # this line is indented (part of the if-block)

# Multiple statements on one line, separated by semicolons (allowed but not idiomatic)
a = 1; b = 2; print(a + b)
````

## 2. Python Variables

- Variables are created the moment you first assign a value — no declaration keyword needed.
- Variable names are case-sensitive, must start with a letter or underscore, and can contain letters, digits, and underscores.
- Python allows multiple assignment and unpacking.

````python
x = 5
y = "Hello, World!"
print(x)
print(y)

# Multiple assignment
a, b, c = 1, 2, 3
print(a, b, c)

# Same value to multiple variables
x = y = z = "Orange"
print(x, y, z)

# Unpacking a list
fruits = ["apple", "banana", "cherry"]
f1, f2, f3 = fruits
print(f1, f2, f3)

# Global vs local scope
def my_func():
    local_var = "I am local"
    print(local_var)

my_func()

# 'global' keyword lets a function modify a global variable
count = 0
def increment():
    global count
    count += 1

increment()
print(count)
````

## 3. Python Data Types

Built-in data types grouped by category:

| Category | Types |
|---|---|

## 4. Python Numbers

Three numeric types: `int`, `float`, `complex`. Python also has a built-in `random` module for random numbers.

````python
x = 1          # int
y = 2.8         # float
z = 1j          # complex

print(type(x), type(y), type(z))

# Arithmetic
print(10 / 3)    # true division -> float
print(10 // 3)   # floor division -> int
print(10 % 3)    # modulus
print(2 ** 8)    # exponent

# Converting between types
a = float(x)
b = int(y)
c = complex(x)
print(a, b, c)

import random
print(random.randrange(1, 10))  # random int between 1 and 9
````

## 5. Python Casting

Explicitly convert (cast) a value from one type to another using `int()`, `float()`, `str()`, `bool()`, etc.

````python
# Casting to int
print(int(1))        # from int
print(int(2.8))       # from float (truncates)
print(int("3"))       # from string

# Casting to float
print(float(1))
print(float("4.2"))

# Casting to str
print(str(2))
print(str(3.0))

# Casting to bool
print(bool(1))    # True
print(bool(0))    # False
print(bool(""))   # False (empty string)
print(bool("a"))  # True (non-empty string)
````

## 6. Python Strings

Strings are sequences of characters, enclosed in single or double quotes. They are **immutable** and support slicing, formatting, and many built-in methods.

````python
s = "Hello, World!"
print(s)
print(s[0])        # indexing
print(s[1:5])      # slicing
print(s[-5:])      # negative slicing
print(len(s))

# Strings are immutable - methods return new strings
print(s.upper())
print(s.lower())
print(s.replace("Hello", "Goodbye"))
print(s.strip())            # remove leading/trailing whitespace
print(s.split(","))         # split into a list

# Concatenation
a = "Hello"
b = "World"
print(a + " " + b)

# Membership
print("World" in s)
print("world" not in s)

# Multiline strings
multiline = """Line 1
Line 2
Line 3"""
print(multiline)

# f-strings (formatted string literals)
name = "Alice"
age = 30
print(f"{name} is {age} years old")

# String testing methods - each returns True/False about the string's content
mystring = "my string"
print(mystring.isalnum())    # False - contains a space, so not ALL chars are alphanumeric
print(mystring.isalpha())    # False - contains a space, so not ALL chars are alphabetic
print(mystring.isdigit())    # False - no digits at all, let alone all-digit
print(mystring.islower())    # True - all cased characters are lowercase
print(mystring.isupper())    # False - no uppercase characters
print(mystring.isspace())    # False - not ALL characters are whitespace
print(mystring.istitle())    # False - not Title Cased (each word capitalized)
print(mystring.startswith("m"))  # True
print(mystring.endswith("a"))    # False
print(mystring.upper())      # 'MY STRING' - new string, original unchanged
print(mystring.lower())      # 'my string' - already lowercase
````

## 7. Python Booleans

Booleans represent one of two values: `True` or `False`. Almost any value is "truthy" except things like `0`, `""`, `[]`, `{}`, and `None`, which are "falsy".

````python
print(10 > 9)     # True
print(10 == 9)    # False
print(10 < 9)     # False

x = "Hello"
y = 15
print(bool(x))    # True - non-empty string
print(bool(y))    # True - non-zero number

# Falsy values
print(bool(False))
print(bool(None))
print(bool(0))
print(bool(""))
print(bool([]))
print(bool({}))

# Functions can return booleans
def is_even(n):
    return n % 2 == 0

print(is_even(4))
print(is_even(7))
````

## 8. Python Operators

- Arithmetic: `+ - * / % ** //`
- Assignment: `= += -= *= /= %= **= //= &= |= ^= >>= <<=`
- Comparison: `== != > < >= <=`
- Logical: `and or not`
- Identity: `is is not`
- Membership: `in not in`
- Bitwise: `& | ^ ~ << >>`

````python
# Arithmetic
print(5 + 3, 5 - 3, 5 * 3, 5 / 3, 5 % 3, 5 ** 3, 5 // 3)

# Assignment
x = 5
x += 3   # same as x = x + 3
print(x)

# Comparison
print(5 == 3, 5 != 3, 5 > 3, 5 < 3)

# Logical
print(True and False, True or False, not True)

# Identity (is checks if two variables point to the SAME object)
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a is b)     # False - different objects with equal value
print(a is c)     # True - same object
print(a == b)     # True - equal value

# Membership
print(3 in [1, 2, 3])
print(4 not in [1, 2, 3])

# Bitwise
print(5 & 3, 5 | 3, 5 ^ 3, ~5, 5 << 1, 5 >> 1)

# Operator precedence: ** > */%// > +- ; use parentheses to be explicit
print(2 + 3 * 4)      # 14
print((2 + 3) * 4)    # 20
````

## 9. Python Lists

Ordered, **mutable**, allow duplicates. Defined with square brackets `[]`.

````python
fruits = ["apple", "banana", "cherry"]
print(fruits)
print(fruits[1])
print(fruits[-1])       # last item
print(fruits[0:2])      # slicing

# Mutating
fruits[1] = "blueberry"
fruits.append("orange")         # add to end
fruits.insert(1, "kiwi")        # insert at index
fruits.remove("cherry")         # remove by value
popped = fruits.pop()           # remove & return last item
print(fruits, popped)

fruits.sort()
print(fruits)
fruits.reverse()
print(fruits)

# Looping
for fruit in fruits:
    print(fruit)

# List comprehension
squares = [x**2 for x in range(6)]
print(squares)

evens = [x for x in range(20) if x % 2 == 0]
print(evens)

# Copying (assignment does NOT copy!)
original = [1, 2, 3]
copy_ = original.copy()   # or list(original)
copy_.append(4)
print(original, copy_)

# Joining lists
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)
print(len(list1))
````

## 10. Python Tuples

Ordered, **immutable**, allow duplicates. Defined with parentheses `()`. Faster than lists and useful when data shouldn't change.

````python
point = (10, 20, 30)
print(point)
print(point[0])
print(point[-1])
print(point[0:2])

# Tuples are immutable - this would raise a TypeError:
# point[0] = 99

# Unpacking
x, y, z = point
print(x, y, z)

# Single-element tuple needs a trailing comma
single = ("apple",)
print(type(single))

# Tuple methods
numbers = (1, 2, 3, 2, 4, 2)
print(numbers.count(2))   # count occurrences
print(numbers.index(3))   # find index of first match

# Looping
for value in point:
    print(value)

# Convert to list, modify, convert back
temp = list(point)
temp.append(40)
point2 = tuple(temp)
print(point2)
````

## 11. Python Sets

Unordered, **mutable**, do **not** allow duplicates. Defined with curly braces `{}`. Great for membership tests and set algebra.

````python
fruits = {"apple", "banana", "cherry", "apple"}  # duplicate "apple" is dropped
print(fruits)

# Gotcha: {} creates an empty DICT, not an empty set - use set() instead
empty_dict = {}
empty_set = set()
print(type(empty_dict), type(empty_set))

# set() also builds a set from any iterable (and de-duplicates it)
from_list = set([1, 2, 2, 3, 3, 3])
from_string = set("banana")   # de-duplicated characters
print(from_list, from_string)

fruits.add("orange")
fruits.update(["mango", "kiwi"])
print(fruits)

fruits.remove("banana")     # raises KeyError if not found
fruits.discard("missing")   # does NOT raise if not found
print(fruits)

# Set algebra
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a.union(b))            # a | b
print(a.intersection(b))     # a & b
print(a.difference(b))       # a - b
print(a.symmetric_difference(b))  # a ^ b

# Membership test (very fast - O(1))
print(3 in a)

# Looping
for fruit in fruits:
    print(fruit)

# Set comprehension
squares = {x**2 for x in range(6)}
print(squares)
````

## 12. Python Dictionaries

Ordered (since Python 3.7+ insertion order is preserved), **mutable**, no duplicate keys. Defined with curly braces as `key: value` pairs.

````python
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}
print(person)
print(person["name"])
print(person.get("age"))
print(person.get("email", "N/A"))  # default if key missing

# Mutating
person["email"] = "john@example.com"   # add
person["age"] = 31                     # update
del person["city"]                     # remove
print(person)

# Keys, values, items
print(list(person.keys()))
print(list(person.values()))
print(list(person.items()))

# Looping
for key, value in person.items():
    print(key, "->", value)

# Membership
print("name" in person)

# Nested dictionaries
family = {
    "child1": {"name": "Emil", "age": 8},
    "child2": {"name": "Tobias", "age": 12},
}
print(family["child2"]["name"])

# Dict comprehension
squares = {x: x**2 for x in range(6)}
print(squares)

# Merging dicts (Python 3.9+)
defaults = {"theme": "dark", "lang": "en"}
overrides = {"lang": "fr"}
merged = defaults | overrides
print(merged)
````

## 13. Python If...Else

Conditional execution using `if`, `elif`, `else`. Python also supports a ternary (conditional expression) shorthand.

````python
a = 33
b = 200

if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")

# Short hand if
if a < b: print("a is less than b")

# Ternary / conditional expression
result = "b is bigger" if b > a else "a is bigger or equal"
print(result)

# Nested conditions
x = 15
if x > 10:
    if x > 20:
        print("x is greater than 20")
    else:
        print("x is between 11 and 20")

# 'and' / 'or' in conditions
age = 25
has_id = True
if age >= 18 and has_id:
    print("Entry allowed")
````

## 14. Python Match (structural pattern matching, Python 3.10+)

`match` / `case` is Python's version of a switch statement, with pattern-matching capabilities beyond simple equality.

````python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case 500 | 502 | 503:            # combine multiple patterns
            return "Server Error"
        case _:                          # default / wildcard
            return "Unknown"

for c in [200, 404, 502, 999]:
    print(c, "->", http_status(c))

# Matching with variable capture and guards
def describe_point(point):
    match point:
        case (0, 0):
            return "Origin"
        case (x, 0):
            return f"On the x-axis at {x}"
        case (0, y):
            return f"On the y-axis at {y}"
        case (x, y) if x == y:
            return "On the diagonal"
        case (x, y):
            return f"Point at ({x}, {y})"

for p in [(0, 0), (5, 0), (0, 5), (3, 3), (2, 7)]:
    print(p, "->", describe_point(p))
````

## 15. Python While Loops

Repeats a block as long as a condition is `True`. Supports `break`, `continue`, and an optional `else` clause (runs when the loop finishes without `break`).

````python
i = 1
while i < 6:
    print(i)
    i += 1

# break
i = 1
while i < 6:
    if i == 4:
        break
    print(i)
    i += 1

# continue
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)

# while...else - else runs if loop completes without a 'break'
i = 1
while i < 4:
    print(i)
    i += 1
else:
    print("i is no longer less than 4")
````

## 16. Python For Loops

Iterates over the items of any sequence (list, tuple, dict, set, string) or other iterable.

````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# Loop through a string
for letter in "banana":
    print(letter)

# range()
for i in range(6):
    print(i)

for i in range(2, 6):
    print(i)

for i in range(2, 30, 3):   # start, stop, step
    print(i)

# break / continue
for fruit in fruits:
    if fruit == "banana":
        break
    print(fruit)

for fruit in fruits:
    if fruit == "banana":
        continue
    print(fruit)

# for...else
for i in range(5):
    print(i)
else:
    print("Loop finished")

# Nested loops
for x in ["red", "blue"]:
    for y in ["small", "big"]:
        print(x, y)

# enumerate - get index + value
for index, fruit in enumerate(fruits):
    print(index, fruit)

# zip - iterate multiple iterables together
names = ["Alice", "Bob"]
ages = [30, 25]
for name, age in zip(names, ages):
    print(name, age)
````

## 17. Python Functions

Defined with `def`. Support default arguments, keyword arguments, `*args`, `**kwargs`, and lambda (anonymous) functions.

````python
def greet(name):
    return f"Hello, {name}!"

print(greet("World"))

# Default parameter value
def greet2(name="Guest"):
    return f"Hello, {name}!"

print(greet2())
print(greet2("Alice"))

# Keyword arguments
def describe_pet(animal, name):
    print(f"I have a {animal} named {name}")

describe_pet(animal="dog", name="Rex")
describe_pet(name="Whiskers", animal="cat")   # order doesn't matter with keywords

# *args - variable number of positional arguments
def total(*numbers):
    return sum(numbers)

print(total(1, 2, 3, 4))

# **kwargs - variable number of keyword arguments
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)

# Return multiple values (as a tuple)
def min_max(numbers):
    return min(numbers), max(numbers)

lo, hi = min_max([4, 2, 9, 1])
print(lo, hi)

# Recursion
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))

# Lambda functions - small anonymous functions
square = lambda x: x ** 2
print(square(5))

add = lambda a, b: a + b
print(add(3, 4))

# Lambdas are often used with map/filter/sorted
nums = [1, 2, 3, 4, 5]
print(list(map(lambda x: x * 2, nums)))
print(list(filter(lambda x: x % 2 == 0, nums)))
print(sorted(["banana", "kiwi", "apple"], key=lambda s: len(s)))
````

## 20. Python Iterators

An iterator is an object with `__next__()` that returns the next value, and raises `StopIteration` when exhausted. Any object with `__iter__()` is *iterable*.

````python
my_list = ["apple", "banana", "cherry"]
my_iter = iter(my_list)

print(next(my_iter))
print(next(my_iter))
print(next(my_iter))
# print(next(my_iter))  # would raise StopIteration

# A for loop uses iterators under the hood
for x in my_list:
    print(x)

# Building a custom iterator class
class Counter:
    def __init__(self, low, high):
        self.current = low
        self.high = high

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.high:
            raise StopIteration
        value = self.current
        self.current += 1
        return value

for num in Counter(1, 5):
    print(num)

# Generators - a simpler way to create iterators
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(5):
    print(num)

gen = count_up_to(3)
print(next(gen), next(gen), next(gen))
````

## 25. Python RegEx

The `re` module provides regular-expression matching, with functions like `search`, `findall`, `split`, and `sub`.

````python
import re

text = "The rain in Spain falls mainly on the plain"

# search - find the first match
match = re.search(r"\bS\w+", text)
print(match.group() if match else None)

# findall - find all matches
matches = re.findall(r"ain", text)
print(matches)

# split - split by pattern
print(re.split(r"\s", text))

# sub - replace matches
print(re.sub(r"ain", "AIN", text))

# Common patterns: validating an email
email_pattern = r"^[\w.+-]+@[\w-]+\.[a-zA-Z]{2,}$"
for candidate in ["user@example.com", "not-an-email"]:
    is_valid = bool(re.match(email_pattern, candidate))
    print(candidate, "->", is_valid)

# Using groups
m = re.search(r"(\d{4})-(\d{2})-(\d{2})", "Date: 2025-12-25")
if m:
    print(m.group(0))   # full match
    print(m.group(1), m.group(2), m.group(3))  # year, month, day
````

## 27. Python Try...Except

Handles errors gracefully using `try` / `except` / `else` / `finally`, instead of crashing the program.

````python
try:
    print(1 / 0)
except ZeroDivisionError:
    print("You can't divide by zero!")

# Catching multiple exception types
try:
    x = int("not a number")
except (ValueError, TypeError) as e:
    print(f"Conversion failed: {e}")

# else - runs if no exception occurred
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error!")
else:
    print(f"Success, result = {result}")

# finally - always runs (cleanup code)
try:
    f = open("nonexistent_file.txt")
except FileNotFoundError:
    print("File not found!")
finally:
    print("This always runs, e.g. for closing resources")

# Raising your own exceptions
def set_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative")
    return age

try:
    set_age(-5)
except ValueError as e:
    print(f"Caught: {e}")

# Custom exception classes
class InsufficientFundsError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(f"Cannot withdraw {amount}, balance is {balance}")
    return balance - amount

try:
    withdraw(100, 150)
except InsufficientFundsError as e:
    print(e)
````

## 28. Python String Formatting

Three main ways to format strings: **f-strings** (recommended, Python 3.6+), `str.format()`, and old-style `%` formatting.

````python
name = "Alice"
age = 30
price = 49.9

# f-strings (preferred)
print(f"{name} is {age} years old")
print(f"Price: ${price:.2f}")        # 2 decimal places
print(f"{age:5d}")                   # width 5, right-aligned
print(f"{name!r}")                   # repr()
print(f"{2 * 21}")                   # expressions allowed inside {}

# str.format()
print("{} is {} years old".format(name, age))
print("{0} is {1}, {0} likes Python".format(name, age))   # positional reuse
print("{n} is {a} years old".format(n=name, a=age))       # named

# Old-style % formatting
print("%s is %d years old" % (name, age))
print("Price: $%.2f" % price)

# Number formatting extras
n = 1234567.891
print(f"{n:,.2f}")     # thousands separator + 2 decimals
print(f"{0.256:.1%}")  # percentage
print(f"{255:x}")      # hexadecimal
print(f"{255:b}")      # binary
````

## 31. Python File Handling

Python's built-in `open()` function reads and writes files. Key modes:

| Mode | Meaning |
|---|---|
| `"r"` | Read (default) — error if the file doesn't exist |
| `"w"` | Write — creates the file if it doesn't exist, **overwrites/truncates** if it does |
| `"a"` | Append — creates the file if it doesn't exist, adds to the end if it does |
| `"x"` | Create — error if the file **already** exists |
| `"t"` | Text mode (default) |
| `"b"` | Binary mode |

Always prefer the `with open(...) as f:` context-manager form — it automatically closes the file (even if an exception occurs), so you never have to remember `f.close()`.

The examples below use Python's `tempfile` module to create a scratch file, so running this notebook doesn't leave files scattered around your project.

````python
import os
import tempfile

# Create an isolated temp directory + file path for this demo
tmp_dir = tempfile.mkdtemp()
file_path = os.path.join(tmp_dir, "demo.txt")
print("Working with:", file_path)
````

### Writing / Creating Files

- `"w"` overwrites the file if it exists (or creates it if it doesn't).
- `"a"` appends to the end without erasing existing content.
- `"x"` fails with `FileExistsError` if the file is already there — useful to avoid accidentally clobbering data.
- `f.write()` writes a string; `f.writelines()` writes a list of strings (it does **not** add newlines for you).

````python
# "w" - write (creates the file, overwrites if it already exists)
with open(file_path, "w") as f:
    f.write("Line 1\n")
    f.write("Line 2\n")

# "a" - append (adds to the end, doesn't erase existing content)
with open(file_path, "a") as f:
    f.write("Line 3\n")

# writelines() writes a list of strings as-is (no automatic newlines)
with open(file_path, "a") as f:
    f.writelines(["Line 4\n", "Line 5\n"])

# "x" - create exclusively; raises FileExistsError if the file already exists
exclusive_path = os.path.join(tmp_dir, "brand_new.txt")
with open(exclusive_path, "x") as f:
    f.write("First and only write allowed here\n")

try:
    with open(exclusive_path, "x") as f:
        f.write("This will fail")
except FileExistsError as e:
    print("FileExistsError:", e)
````

### Reading Files

- `f.read()` — reads the entire file as a single string (optionally pass a max number of characters).
- `f.readline()` — reads a single line, including its trailing `\n`.
- `f.readlines()` — reads all lines into a list of strings.
- Looping directly over the file object (`for line in f:`) reads it **line by line** and is the most memory-efficient way to process large files.

````python
with open(file_path, "r") as f:
    content = f.read()
print(content)

with open(file_path, "r") as f:
    first_10_chars = f.read(10)
print(repr(first_10_chars))

with open(file_path, "r") as f:
    first_line = f.readline()
    second_line = f.readline()
print(repr(first_line), repr(second_line))

with open(file_path, "r") as f:
    all_lines = f.readlines()
print(all_lines)

# Most memory-efficient way to process large files: iterate line by line
with open(file_path, "r") as f:
    for line in f:
        print(line.strip())   # .strip() removes the trailing newline
````

### Checking Existence, Deleting Files & Folders

- Always check `os.path.exists()` before deleting, to avoid a `FileNotFoundError`.
- `os.remove(path)` deletes a file.
- `os.rmdir(path)` deletes an **empty** folder.
- `shutil.rmtree(path)` deletes a folder and everything inside it (use with care — this is not reversible).

````python
import shutil

print(os.path.exists(file_path))

if os.path.exists(file_path):
    os.remove(file_path)
    print("Deleted:", file_path)

print(os.path.exists(file_path))

# Deleting a non-existent file without checking first raises FileNotFoundError
try:
    os.remove(file_path)
except FileNotFoundError as e:
    print("FileNotFoundError:", e)

# Clean up the whole temp directory (folder + any remaining files inside it)
shutil.rmtree(tmp_dir)
print("Directory removed:", not os.path.exists(tmp_dir))
````

### Working with Paths

The `os.path` module (or the more modern `pathlib`) helps build paths portably across operating systems instead of hardcoding `/` or `\\`.

````python
from pathlib import Path

p = Path(tempfile.mkdtemp()) / "notes" / "todo.txt"
p.parent.mkdir(parents=True, exist_ok=True)   # create intermediate folders
p.write_text("Buy milk\nWalk the dog\n")      # pathlib shorthand for open()+write()

print(p.read_text())
print(p.exists(), p.name, p.suffix, p.parent)

# Clean up
shutil.rmtree(p.parent.parent)
````

## 32. Python VirtualEnv

A virtual environment is an isolated Python environment with its own installed packages, so different projects don't conflict with each other. This is a shell workflow (not run inside notebook code cells).

```bash
# Create a virtual environment named "venv"
python3 -m venv venv

# Activate it
source venv/bin/activate        # macOS / Linux
venv\Scripts\activate           # Windows

# Confirm you're using the venv's python/pip
which python
which pip

# Install packages into the isolated environment
pip install requests pandas

# Save dependencies for others / other machines
pip freeze > requirements.txt

# Recreate the same environment elsewhere
pip install -r requirements.txt

# Deactivate when done
deactivate
```

Tips:
