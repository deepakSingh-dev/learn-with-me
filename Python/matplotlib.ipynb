# Matplotlib Tutorial

A hands-on, runnable reference for Matplotlib — Python's foundational data-visualization library. Run each code cell as you go; every plot renders inline. Feel free to tweak values and re-run.

```python
# pip install matplotlib   (uncomment to install)

import matplotlib.pyplot as plt
import numpy as np

print(plt.matplotlib.__version__)
```

## 1. Matplotlib Introduction

- Matplotlib is Python's most widely used plotting library — the foundation that pandas' `.plot()`, seaborn, and many other visualization tools are built on top of.
- The `pyplot` module (`matplotlib.pyplot`, conventionally imported as `plt`) provides a MATLAB-style interface: build a figure step by step with function calls, then display it.
- A **Figure** is the whole window/canvas; an **Axes** is a single plot within it (a figure can contain multiple Axes via subplots).

```python
# The simplest possible plot: pyplot keeps track of the "current" figure/axes for you
plt.plot([1, 2, 3, 4])   # a single list is treated as y-values; x defaults to 0, 1, 2, 3...
plt.show()
plt.close()
```

## 2. Matplotlib Getting Started

Pass explicit `x` and `y` arrays to `plt.plot()` to control both axes. NumPy arrays, Python lists, and pandas Series all work.

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
plt.close()

# Plotting just two points draws a single straight line between them
plt.plot(np.array([1, 8]), np.array([3, 10]))
plt.show()
plt.close()

# A single point with no line: use a marker and no linestyle
plt.plot(np.array([3]), np.array([5]), "o")
plt.show()
plt.close()
```

## 3. Matplotlib Pyplot

- `plt.plot(x, y)` draws a line through the given points, in order.
- Without an explicit `x`, the point indices `0, 1, 2, ...` are used automatically.
- Multiple `plot()` calls before a single `show()` draw multiple lines on the same Axes.

```python
# x defaults to the index positions
ypoints = np.array([3, 8, 1, 10, 5])
plt.plot(ypoints)
plt.show()
plt.close()

# Multiple lines on one plot
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
plt.close()
```

## 4. Matplotlib Markers

Use the `marker` parameter (or a shorthand format string) to emphasize each data point. Common markers: `'o'` circle, `'*'` star, `'s'` square, `'x'`/`'X'` X, `'^'` triangle, `'D'` diamond.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker="o")
plt.show()
plt.close()

plt.plot(ypoints, marker="*")
plt.show()
plt.close()

# Format string shorthand: 'marker|line|color', e.g. "o:r" = circle markers, dotted line, red
plt.plot(ypoints, "o:r")
plt.show()
plt.close()

# Customize marker size/color independently from the line
plt.plot(ypoints, marker="o", ms=15, mec="green", mfc="hotpink")
plt.show()
plt.close()
```

## 5. Matplotlib Line

- `linestyle` (or `ls`): `'-'` solid (default), `'--'` dashed, `':'` dotted, `'-.'` dash-dot.
- `color` (or `c`): a name, hex code, or shorthand letter (`'r'`, `'g'`, `'b'`, ...).
- `linewidth` (or `lw`): line thickness.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle="dotted")
plt.show()
plt.close()

plt.plot(ypoints, ls="--", color="#4CAF50", linewidth="5")
plt.show()
plt.close()

# Multiple styled lines together
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1, color="r")
plt.plot(y2, color="b")
plt.show()
plt.close()
```

## 6. Matplotlib Labels and Title

`plt.xlabel()`, `plt.ylabel()`, and `plt.title()` annotate the plot. `fontdict` customizes font styling, and `loc` positions the title (`'left'`, `'center'`, `'right'`).

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.show()
plt.close()

# Custom font styling and title alignment
font1 = {"family": "serif", "color": "blue", "size": 18}
font2 = {"family": "serif", "color": "darkred", "size": 12}

plt.plot(x, y)
plt.title("Sports Watch Data", fontdict=font1, loc="left")
plt.xlabel("Average Pulse", fontdict=font2)
plt.ylabel("Calorie Burnage", fontdict=font2)
plt.show()
plt.close()
```

## 7. Matplotlib Grid

`plt.grid()` adds gridlines. Use `axis="x"` or `axis="y"` to restrict which gridlines appear, and pass standard line-styling kwargs to customize their look.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.show()
plt.close()

# Only vertical gridlines, styled
plt.plot(x, y)
plt.grid(axis="x", color="green", linestyle="--", linewidth=0.5)
plt.show()
plt.close()
```

## 8. Matplotlib Subplots

- `plt.subplot(rows, cols, index)` places multiple plots in one figure, one at a time (index is 1-based).
- `plt.subplots(rows, cols)` is the more modern approach — it returns a `(figure, axes)` pair up front, letting you address each Axes directly (recommended for anything beyond a trivial grid).

```python
# plt.subplot(): build a 1x2 grid, one call per plot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)   # 1 row, 2 cols, this is plot #1
plt.plot(x, y)
plt.title("Plot 1")

y2 = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)   # 1 row, 2 cols, this is plot #2
plt.plot(x, y2)
plt.title("Plot 2")

plt.suptitle("plt.subplot() Example")
plt.show()
plt.close()

# plt.subplots(): returns (figure, axes) - the modern, recommended pattern
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes[0, 0].plot(x, y)
axes[0, 0].set_title("Top Left")
axes[0, 1].scatter(x, y2)
axes[0, 1].set_title("Top Right")
axes[1, 0].bar(x, y)
axes[1, 0].set_title("Bottom Left")
axes[1, 1].plot(x, y, "o--r")
axes[1, 1].set_title("Bottom Right")

fig.tight_layout()   # prevents titles/labels from overlapping
plt.show()
plt.close()
```

## 9. Matplotlib Scatter

`plt.scatter(x, y)` plots individual points without connecting them — ideal for showing the relationship (or correlation) between two variables. `c` and `s` control per-point color and size, and `cmap` maps color values to a colormap.

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)
plt.show()
plt.close()

# Two datasets on one scatter plot, each with its own color
x1 = np.array([89, 43, 36, 36, 95, 10, 66, 34, 38, 20])
y1 = np.array([21, 46, 3, 35, 67, 95, 53, 72, 58, 10])
x2 = np.array([26, 29, 48, 64, 6, 5, 36, 66, 72, 40])
y2 = np.array([26, 34, 90, 33, 38, 20, 56, 2, 47, 15])

plt.scatter(x1, y1, color="hotpink", label="group 1")
plt.scatter(x2, y2, color="#88c999", label="group 2")
plt.legend()
plt.show()
plt.close()

# Color gradient + size per point, with a colorbar
np.random.seed(42)
x3 = np.random.randint(100, size=100)
y3 = np.random.randint(100, size=100)
colors = np.random.randint(100, size=100)
sizes = np.random.randint(10, 300, size=100)

plt.scatter(x3, y3, c=colors, s=sizes, cmap="viridis", alpha=0.6)
plt.colorbar()
plt.show()
plt.close()
```

## 10. Matplotlib Bars

`plt.bar()` draws vertical bars, `plt.barh()` draws horizontal ones. Customize with `color` and `width` (or `height` for `barh`).

```python
categories = np.array(["A", "B", "C", "D"])
values = np.array([3, 8, 1, 10])

plt.bar(categories, values)
plt.show()
plt.close()

plt.barh(categories, values, color="hotpink")
plt.show()
plt.close()

plt.bar(categories, values, color="#4CAF50", width=0.4)
plt.show()
plt.close()

# Grouped bars: offset x positions for side-by-side comparison
x = np.arange(len(categories))
values_2022 = np.array([3, 8, 1, 10])
values_2023 = np.array([5, 6, 4, 9])
width = 0.35

plt.bar(x - width / 2, values_2022, width, label="2022")
plt.bar(x + width / 2, values_2023, width, label="2023")
plt.xticks(x, categories)
plt.legend()
plt.show()
plt.close()
```

## 11. Matplotlib Histograms

`plt.hist()` bins a dataset of raw values and plots the count in each bin — useful for visualizing a **distribution**, unlike `bar()` which plots pre-aggregated categories.

```python
np.random.seed(42)
x = np.random.normal(170, 10, 250)   # 250 samples ~ normal distribution, mean=170, std=10

plt.hist(x)
plt.show()
plt.close()

# Customize bin count and styling
plt.hist(x, bins=20, color="skyblue", edgecolor="black")
plt.title("Distribution of Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Count")
plt.show()
plt.close()
```

## 12. Matplotlib Pie Charts

`plt.pie()` shows each value's proportion of the whole. Slices start at the 3 o'clock position and go counter-clockwise by default; `startangle` and `explode` adjust the layout.

```python
values = np.array([35, 25, 25, 15])
labels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(values, labels=labels)
plt.show()
plt.close()

# Start angle, "explode" one slice out, add a shadow, and add a legend
explode_amounts = [0.2, 0, 0, 0]   # only the first slice ("Apples") is pulled out
plt.pie(
    values,
    labels=labels,
    startangle=90,
    explode=explode_amounts,
    shadow=True,
    autopct="%1.1f%%",   # show percentage on each slice
)
plt.legend(title="Fruits")
plt.show()
plt.close()
```

## 13. Matplotlib Legends

`plt.legend()` labels each series so a reader can tell them apart — pass `label=` to each plotting call, then call `legend()` once. `loc` controls placement.

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, label="Line 1")
plt.plot(y2, label="Line 2")
plt.legend()
plt.show()
plt.close()

# Custom legend location and title
plt.plot(y1, label="Line 1")
plt.plot(y2, label="Line 2")
plt.legend(loc="upper center", title="Legend")
plt.show()
plt.close()
```

## 14. Matplotlib Saving Figures

`plt.savefig()` writes the current figure to disk (PNG, PDF, SVG, ...) instead of (or in addition to) displaying it — call it **before** `plt.show()`, since `show()` can clear the figure afterward. `dpi` controls resolution.

```python
import tempfile, os

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("Sine Wave")

out_path = os.path.join(tempfile.mkdtemp(), "sine_wave.png")
plt.savefig(out_path, dpi=150, bbox_inches="tight")
plt.show()
plt.close()

print("Saved to:", out_path)
print("File exists:", os.path.exists(out_path), "-", os.path.getsize(out_path), "bytes")
```

## 15. Matplotlib Styles

`plt.style.use()` applies a complete visual theme (colors, fonts, gridlines) to every subsequent plot. `plt.style.available` lists all built-in styles. Use `with plt.style.context(...)` to apply a style temporarily without affecting later cells.

```python
print(plt.style.available[:8])   # a preview of the available built-in styles

x = np.linspace(0, 10, 100)

# Apply a style just for one plot, then automatically revert
with plt.style.context("ggplot"):
    plt.plot(x, np.sin(x), label="sin(x)")
    plt.plot(x, np.cos(x), label="cos(x)")
    plt.legend()
    plt.title("ggplot style")
    plt.show()
    plt.close()

# Back to default style here
plt.plot(x, np.sin(x))
plt.title("default style")
plt.show()
plt.close()
```

## 16. Matplotlib Text and Annotations

- `plt.text(x, y, s)` places arbitrary text at a data coordinate.
- `plt.annotate()` does the same but can also draw an arrow pointing from the text to a specific point — useful for calling out a notable feature in the data.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

# Plain text label
plt.text(1, 0.8, "Peak region", fontsize=10, color="darkred")

# Annotation with an arrow pointing at the actual peak
peak_idx = np.argmax(y)
plt.annotate(
    "Maximum",
    xy=(x[peak_idx], y[peak_idx]),        # the point being annotated
    xytext=(x[peak_idx] + 1, y[peak_idx] - 0.3),  # where the text goes
    arrowprops=dict(facecolor="black", arrowstyle="->"),
)
plt.show()
plt.close()
```

## 17. Matplotlib Colormaps and `imshow`

`plt.imshow()` renders a 2-D array as an image, mapping values to colors via a `cmap` (colormap) — used both for real images and for visualizing 2-D data like heatmaps or correlation matrices.

```python
np.random.seed(0)
data = np.random.rand(10, 10)

plt.imshow(data, cmap="viridis")
plt.colorbar(label="value")
plt.title("Random Heatmap")
plt.show()
plt.close()

# A correlation-style matrix is a very common real-world use of imshow
matrix = np.array([
    [1.0, 0.8, 0.2],
    [0.8, 1.0, 0.5],
    [0.2, 0.5, 1.0],
])
plt.imshow(matrix, cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(3), ["A", "B", "C"])
plt.yticks(range(3), ["A", "B", "C"])
plt.title("Correlation Matrix")
plt.show()
plt.close()
```

## 18. Matplotlib 3D Plotting

Pass `projection="3d"` when creating an Axes to unlock 3D plotting: `plot_surface()`, `scatter()`, and `plot()` all work in three dimensions.

```python
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(projection="3d")

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

surf = ax.plot_surface(X, Y, Z, cmap="viridis")
fig.colorbar(surf, shrink=0.5)
ax.set_title("3D Surface Plot")
plt.show()
plt.close()

# 3D scatter
fig = plt.figure(figsize=(6, 5))
ax = fig.add_subplot(projection="3d")
np.random.seed(1)
xs = np.random.rand(50)
ys = np.random.rand(50)
zs = np.random.rand(50)
ax.scatter(xs, ys, zs, c=zs, cmap="plasma")
ax.set_title("3D Scatter Plot")
plt.show()
plt.close()
```
